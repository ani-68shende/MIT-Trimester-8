# Learning how to identify and remove duplicate values

```python
data.duplicated(subset=None, keep=False).value_counts()
```

```python
boolean_values = data.duplicated(subset=None, keep=False)
```

```python
boolean_values
```

### Passing negation of bool_values so as to get the unique values

```python
unique_values = data[~boolean_values]
```

```python
unique_values
```

## Using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
#https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv
data1 = pd.read_csv("https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv",header = None, usecols =[0,1,2],names=['class', 'alcohol', 'malic_acid'])
print(f'Shape: {data1.shape}')
data1.head(n=3)
```

```python
data1
```

```python
data1.shape
```

```python
scaling = MinMaxScaler()
```

```python
scaling.fit_transform(data1[['alcohol', 'malic_acid']])
```

```python
standard_scaler = StandardScaler()
standard_scaler.fit_transform(data1[['alcohol', 'malic_acid']])
```
